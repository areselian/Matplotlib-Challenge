```python
# Dependencies and Setup
%matplotlib inline
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import os
os.getcwd()
os.listdir(os.getcwd())

# Hide warning messages in notebook
import warnings
warnings.filterwarnings('ignore')

# File to Load (Remember to Change These)
mouse_drug_data_to_load = "mouse_drug_data.csv"
clinical_trial_data_to_load = "clinicaltrial_data.csv"

# Read the Mouse and Drug Data and the Clinical Trial Data
mouse_drug_data= pd.read_csv("mouse_drug_data.csv")
clinical_trial_data=pd.read_csv("clinicaltrial_data.csv")

# Combine the data into a single dataset
combined_data_df = pd.merge(clinical_trial_data, mouse_drug_data,how="left", on="Mouse ID")

# Display the data table for preview

combined_data_df.head(10)
```

## Tumor Response to Treatment

```python
combined_data_df.columns
```

```python
# Store the Mean Tumor Volume Data Grouped by Drug and Timepoint 

dt_vol=combined_data_df.iloc[:,[1,2,4]]
dt_vol

drug_timepoint_grouped = dt_vol.groupby(['Drug', "Timepoint"]).mean()
drug_timepoint_grouped.head()

# Convert to DataFrame
tumor_vol_df = drug_timepoint_grouped.add_suffix('_Mean').reset_index()
#tumor_vol_df = pd.DataFrame(tumor_vol_df)

# Preview DataFrame
tumor_vol_df.head()
```

```python
# Store the Standard Error of Tumor Volumes Grouped by Drug and Timepoint
# Convert to DataFrame
# Preview DataFrame

sem_grouped = dt_vol.groupby(["Drug", "Timepoint"]).sem()
sem_grouped

sem_df = sem_grouped.add_suffix('_Error').reset_index()
sem_df.head()

sem_df = pd.DataFrame(sem_df)
sem_df.head()
```

```python
# Minor Data Munging to Re-Format the Data Frames - Use Pivot Table for Averages
#df.pivot(index='foo', columns='bar', values='baz')
# Preview that Reformatting worked

#Format Mean 

tumor_mean_df=tumor_vol_df.pivot(index='Timepoint', columns='Drug', values='Tumor Volume (mm3)_Mean')
tumor_mean_df.head()

new_tumor_mean_df = tumor_mean_df.loc[:,["Capomulin", "Infubinol", "Ketapril", "Placebo"]]
new_tumor_mean_df
```

```python
# Minor Data Munging to Re-Format the Data Frames - Use Pivot Table for Sems
# Preview that Reformatting worked
tumor_error_df = sem_df.pivot(index='Timepoint', columns='Drug', values='Tumor Volume (mm3)_Error')
tumor_error_df.head()

new_tumor_error_df = tumor_error_df.loc[:,["Capomulin", "Infubinol", "Ketapril", "Placebo"]]
new_tumor_error_df
```

```python
#Set the four drugs for plotting 
#Scatter plot needs Error Bars, proper Labeling (titles, axes labels, legend labels,x/y axis limits, etc.)
#https://matplotlib.org/3.1.1/api/markers_api.html - Markers for Plots

four_drugs_df = tumor_error_df.loc[:,["Capomulin", "Infubinol", "Ketapril", "Placebo"]]
four_drugs_df
```

```python
# Generate the Plot (with Error Bars)


#ax1= plt.scatter(new_met_average_df.index.values, new_met_average_df['Capomulin'], color="g", label= "Capomulin")
#ax1_error= plt.errorbar(new_met_average_df.index.values, new_met_average_df['Capomulin'],yerr=new_sem_met_error_df['Capomulin'], color="g", label= None, linestyle='None')

#fig, axes = plt.subplots() - https://www.southampton.ac.uk/~fangohr/training/python/notebooks/Matplotlib.html
#plt.errorbar(trip.index, trip['gas'], yerr=trip['std'])
#plt.plot(range(10), linestyle='--', marker='o', color='b')
#plt.plot(range(10), '--bo')


fig, axes = plt.subplots(figsize=(8,6))

ax1 = plt.scatter(new_tumor_mean_df.index.values, new_tumor_mean_df["Capomulin"], marker ="o", color='r',label="Capomulin")
ax1 = plt.errorbar(new_tumor_mean_df.index.values, new_tumor_mean_df["Capomulin"], yerr=new_tumor_error_df["Capomulin"], color='r', label= None, linestyle='--')

ax2 = plt.scatter(new_tumor_mean_df.index.values, new_tumor_mean_df["Infubinol"], marker ="s", color='b', label="Infubinol", linestyle='--')
ax2 = plt.errorbar(new_tumor_mean_df.index.values, new_tumor_mean_df["Infubinol"], yerr=new_tumor_error_df["Infubinol"], color='b', label= None, linestyle='--')

ax3 = plt.scatter(new_tumor_mean_df.index.values, new_tumor_mean_df["Ketapril"], marker ="P", color='g', label="Ketapril", linestyle='--')
ax3 = plt.errorbar(new_tumor_mean_df.index.values, new_tumor_mean_df["Ketapril"], yerr=new_tumor_error_df["Ketapril"], color='g', label= None, linestyle='--')

ax4 = plt.scatter(new_tumor_mean_df.index.values, new_tumor_mean_df["Placebo"], marker ="*", color='k', label="Placebo", linestyle='--')
ax4 = plt.errorbar(new_tumor_mean_df.index.values, new_tumor_mean_df["Placebo"], yerr=new_tumor_error_df["Placebo"], color='k', label= None, linestyle='--')

# how the tumor volume changes over time for each treatment.

#HOW TO INCLUDE LINE STYLING IN LEGEND? (handles= [ax1, ax2, ax3, ax4])

plt.legend(labels=four_drugs_df, loc='best')

plt.title("Tumor Response to Treatment",fontsize=12, fontweight='bold' )
plt.xlabel("Treatment Timepoint in Days", fontsize=11, fontweight='bold')
plt.ylabel("Tumor Volume Mean", fontsize=11, fontweight='bold')
plt.xlim(-3,50)
plt.ylim(30,75)
plt.grid()

# Save the Figure
plt.savefig("tumor_response_to_treatment.png")

# Show the Figure
plt.show()
```

## Metastatic Response to Treatment

```python
# Store the Mean Met. Site Data Grouped by Drug and Timepoint 
mean_met = combined_data_df.iloc[:,[4,1,3]]
mean_met

mean_met_grouped = mean_met.groupby(['Drug', "Timepoint"]).mean()
mean_met_grouped

# Convert to DataFrame
# Preview DataFrame
met_df=mean_met_grouped.add_suffix('_Mean').reset_index()
met_df.head()

met_df=pd.DataFrame(met_df)
```

```python
# Store the Standard Error associated with Met. Sites Grouped by Drug and Timepoint 
sem_met = mean_met.groupby(["Drug", "Timepoint"]).sem()
sem_met

sem_met_df = sem_met.add_suffix('_Error').reset_index()
sem_met_df.head()

# Convert to DataFrame
# Preview DataFrame

sem_met_df = pd.DataFrame(sem_met_df)
sem_met_df.head()
```

```python
# Minor Data Munging to Re-Format the Data Frames

mean_met_df=met_df.pivot(index='Timepoint', columns='Drug', values='Metastatic Sites_Mean')
mean_met_df.head()

# Preview that Reformatting worked
new_mean_met_df = mean_met_df.loc[:,["Capomulin", "Infubinol", "Ketapril", "Placebo"]]
new_mean_met_df
```

## Survival Rates

```python
# Store the Count of Mice Grouped by Drug and Timepoint (W can pass any metric)

combined_data_df.head()
mouse_df = combined_data_df.iloc[:,[0,1,4]]
mouse_df

mouse_grouped_df = mouse_df.groupby(["Drug", "Timepoint"]).count()
mouse_grouped_df


# Convert to DataFrame
# Preview DataFrame


new_mouse_df = mouse_grouped_df.add_suffix('_Counts').reset_index()
new_mouse_df.head()

```

```python
# Minor Data Munging to Re-Format the Data Frames
final_mouse_df=new_mouse_df.pivot(index='Timepoint', columns='Drug', values='Mouse ID_Counts')
final_mouse_df.head()

final_mouse_df2 = final_mouse_df.loc[:,["Capomulin", "Infubinol", "Ketapril","Placebo"]]
final_mouse_df2.head()

# Preview the Data Frame
final_mouse_df2
```

```python
#Percentages
final_mouse_df2["Capomulin Survival"]=final_mouse_df2["Capomulin"]/25*100
final_mouse_df2["Infubinol Survival"]=final_mouse_df2["Infubinol"]/25*100
final_mouse_df2["Ketapril Survival"]=final_mouse_df2["Ketapril"]/25*100
final_mouse_df2["Placebo Survival"]=final_mouse_df2["Placebo"]/25*100

final_mouse_df2
```

```python
survival_rate_df = final_mouse_df2.loc[:,["Capomulin Survival", 
                                          "Infubinol Survival", 
                                          "Ketapril Survival", 
                                          "Placebo Survival"]]
survival_rate_df




```

```python
# Generate the Plot (Accounting for percentages)

fig, axes = plt.subplots(figsize=(8,6))

ax1 = plt.scatter(survival_rate_df.index.values, survival_rate_df["Capomulin Survival"], marker ="o", color='r', label="Capomulin",linestyle="--")

ax2 = plt.scatter(survival_rate_df.index.values, survival_rate_df["Infubinol Survival"], marker ="s", color='b', label="Infubinol", linestyle="--")

ax3 = plt.scatter(survival_rate_df.index.values, survival_rate_df["Ketapril Survival"], marker ="P", color='g', label="Ketapril", linestyle="--")

ax4 = plt.scatter(survival_rate_df.index.values, survival_rate_df["Placebo Survival"], marker ="*", color='k', label="Placebo", linestyle="--")



plt.legend(labels=four_drugs_df, loc='lower left')

plt.title("Survival During Treatment",fontsize=12, fontweight='bold' )
plt.xlabel("Treatment Timepoint in Days", fontsize=11, fontweight='bold')
plt.ylabel("Survival Rate (%)", fontsize=11, fontweight='bold')
plt.xlim(-5,50)
plt.ylim(30,110)
plt.grid()

# Save the Figure
plt.savefig("survival_during_treatment")

# Show the Figure
plt.show()
```
